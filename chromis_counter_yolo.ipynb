```python
from ultralytics import YOLO
from ultralytics.solutions import object_counter
import cv2

# Load the YOLO model
model = YOLO("./runs/detect/train5/weights/best.pt")

# Open video capture
cap = cv2.VideoCapture("/Users/jenny/OneDrive/Desktop/chromis/dataset/C0058.mp4")
assert cap.isOpened(), "Error reading video file"
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))

# Setup video writer
video_writer = cv2.VideoWriter("/Users/jenny/OneDrive/Desktop/chromis/AnnotatedVideos/C0058_Counted.mp4",
                               cv2.VideoWriter_fourcc(*'mp4v'),
                               fps,
                               (w, h))

# Initialize tracking data
last_centroids = {}  # Tracks centroids of fishes in the last frame
id_counter = 0
max_distance = 50  # Max distance a fish can move per frame


while cap.isOpened():
    success, im0 = cap.read()
    if not success:
        break
    
    # Detect and track objects
    results = model.track(im0, persist=True, show=False)
    
    current_centroids = {}
    # Process detections
    if results[0].boxes.xyxy[0].shape[0] > 0:  # Ensure there are detections
        print(results[0].boxes.xyxy[0])
        for bbox in results[0].boxes.xyxy:
            x1, y1, x2, y2 = bbox[0].item(), bbox[1].item(), bbox[2].item(), bbox[3].item()
            centroid = ((x1 + x2) // 2, (y1 + y2) // 2)
            
            # Match current detection with previous detections
            matched = False
            for last_id, last_centroid in last_centroids.items():
                if np.linalg.norm(np.array(centroid) - np.array(last_centroid)) < max_distance:
                    current_centroids[last_id] = centroid
                    matched = True
                    break
            
            if not matched:
                current_centroids[id_counter] = centroid
                id_counter += 1
            
            # Optional: Visualize tracking
            cv2.rectangle(im0, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(im0, f"ID: {list(current_centroids.keys())[-1]}", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
    
    # Update centroids for next frame
    last_centroids = current_centroids.copy()
    
    video_writer.write(im0)

# Cleanup
cap.release()
video_writer.release()
cv2.destroyAllWindows()
```

```python
import logging
import cv2
from ultralytics import YOLO

# Set logging level to ERROR to suppress the detailed output from YOLO
logging.getLogger().setLevel(logging.ERROR)

# Load the video
video_path = "C:/Users/jenny/OneDrive/Desktop/chromis/dataset/C0058.MP4"
cap = cv2.VideoCapture(video_path)
model = YOLO("./runs/detect/train5/weights/best.pt")

max_fishes = 0
frame_counter = 0

# Process video frames
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break


    # Predict using the model
    results = model.predict(frame)
    num_fishes = len(results[0])
    
    # Update the maximum number of fishes found
    if num_fishes > max_fishes:
        max_fishes = num_fishes
        print(f"Frame {frame_counter}: Maximum number of fishes updated to: {max_fishes}")

    frame_counter += 1  # Increment frame counter

    # Optional: Skip frames to increase processing speed
    for _ in range(4):  # Skip the next 4 frames
        ret = cap.grab()
        if not ret:
            break

# Release the video capture object
cap.release()

print(f"Maximum number of fishes detected in any frame: {max_fishes}")
```
